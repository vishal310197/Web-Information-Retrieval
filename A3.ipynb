```python
def precisionrecall(seq):
    pre = []
    re = []
    f1 = []
    position = []
    short_p = []
    short_r = []
    short_f1 = []
    seqlen = 50
    rel = 15
    tp = 0
    for i in seq:
        if("TP" in i):
            tp+=1
            position.append(seq.index(i)+1)
        precision = tp/(seq.index(i)+1)
        pre.append(precision)
        recall = tp/rel
        re.append(recall)
        f = (2*precision*recall)/(precision+recall)
        f1.append(f)
        
    print(pre)
    print(re)
    print(f1)
    short_p = list(map(lambda x: round(x,2), pre))
    short_r = list(map(lambda x: round(x,2), re))
    short_f1 = list(map(lambda x: round(x,2), f1))
    print(short_p)
    print(short_r)
    print(short_f1)
    print(sum(pre))
    print(sum(re))
    
    rank = range(len(pre))
    plt.plot(rank,pre)
    plt.plot(rank,re)
    plt.show()
```
