```python
D0="cats runs behind rats"
D1="dogs runs behind cats"
D2="rats runs cats"
D3="behind runs cats dogs"
q="behind rats"

corpus=[D0,D1,D2,D3,q]
word_bag= set(D0.split()+D1.split()+D2.split()+D3.split()+q.split())


def term_frequency(word_bag, corpus):
    lis = []
    for word in word_bag:
        for match in corpus:
            m = match.count(word)
            lis.append(m)
    lis2 = []
   
    for i in range(0,len(corpus)):
         lis2.append(lis[i::len(corpus)])
    lis2 = np.array(lis2)
    lis2 = lis2.transpose()
    return(pd.DataFrame(lis2,index = word_bag, columns = corpus))
        

term_frequency(word_bag, corpus)
```

```python
dff = term_frequency(word_bag, corpus)
```

```python
lq = []
c = 0
for i in word_bag:
    for j in corpus:
        if i in j:
            c = c+1
    lq.append(c)
    c = 0
print(lq)
dff['df'] = np.array(lq)
dff['global_weight'] = np.log10((len(corpus)/dff['df']))
```
