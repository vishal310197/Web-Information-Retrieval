```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('flower.csv')
df
```

```python
confusion_matrix = pd.crosstab(df['gold_standard'], df['retrieved'], rownames=['Actual'], colnames=['Predicted'])
confusion_matrix
```

```python
#accuracy
np.sum(np.diag(confusion_matrix))/np.sum(np.sum(confusion_matrix))
```
