```python
#relevnt docs
df2 = df[df['Document'].str.contains("Data Science")]
df2['Document ID']
```

```python
#indexing
df2['index'] = range(1,len(df2)+1)
df['index'] = range(1,len(df)+1)
df
```

```python
#relevant or not
df['new'] = np.where(df['Document'].str.contains("Data Science"), 1, 0)
df
```

```python
df
```

```python
i = []
k = 0
for col in df['new']:
    k = k + col
    i.append(k)
print(i)

```

```python
#relevance instance
df['new_col'] = np.array(i)
```

```python
df
```

```python
#calculate precision
df['precision'] = df['new_col']/df['index']
```

```python
df
```

```python
#precision sum
print(df['precision'].sum())
```

```python
#calculate recall
df['recall'] = df['new_col']/15
df
```

```python
print(df['recall'].sum())
```

```python
df['f1'] = (2*df['precision']*df['recall'])/(df['precision']+df['recall'])
df
```
