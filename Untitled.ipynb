```python

import pandas as pd
import numpy as np
d0 = "preliminary findings in corona research"
d1 = "novel corona research findings"
d2 = "new research to corona healing"
d3 = "healing novel corona research"

corpus = [d0, d1, d2, d3]
vocab = set(d0.split() + d1.split() + d2.split() + d3.split())
vocab
```

```python
def term_frequency(vocab):
    d0_dict = dict.fromkeys(vocab,0)
    d1_dict = dict.fromkeys(vocab,0)
    d2_dict = dict.fromkeys(vocab,0)
    d3_dict = dict.fromkeys(vocab,0)
    
    for word in d0.split():
        d0_dict[word]+=1
    for word in d1.split():
        d1_dict[word]+=1
    for word in d2.split():
        d2_dict[word]+=1
    for word in d3.split():
        d3_dict[word]+=1
    
    return pd.DataFrame([d0_dict,d1_dict,d2_dict,d3_dict], index = ['d0','d1','d2','d3']).transpose()
```

```python
tf_idf = term_frequency(vocab)
tf_idf
```

```python
def term_frequencyy(vocab, corpus):
    df = pd.DataFrame(index = list(vocab),columns=["d0","d1","d2","d3"])
    
    for doc,col in zip(corpus,df.columns):
        for word in corpus:
            df[col][word] = doc.count(word)
    return df
```

```python
df = term_frequencyy(vocab,corpus)
df
```

```python
def document_frequency(corpus,vocab):
    c = 0
    doc_freq = []
    for i in vocab:
        for j in corpus:
            if(i in j):
                c+=1
        doc_freq.append(c)
        c=0
    return doc_freq
```

```python
doc_freq = document_frequency(corpus,vocab)
doc_freq
```
